# Data cleaning  

```python
df=df.drop_duplicates()
df.duplicated().sum()
```

```python
df['gender']=df['gender'].fillna(df['gender'].mode()[0])
df['age']=df['age'].fillna(df['age'].mean())
df['heart_disease']=df['heart_disease'].fillna(df['heart_disease'].mode()[0])
df['bmi']=df['bmi'].fillna(df['bmi'].mean())
df['HbA1c_level']=df['HbA1c_level'].fillna(df['HbA1c_level'].mean())
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
import numpy as np
```

```python
a=['age','hypertension','heart_disease',
   'bmi','HbA1c_level','blood_glucose_level']
for i in a:
    
    q1=df[i].quantile(.25)
    q3=df[i].quantile(.75)
    
    iqr=q3-q1
    
    lb=q1-(1.5*iqr)
    ub=q3+(1.5*iqr)
    
    df[i]=np.where(df[i]<lb,lb,
                   np.where(df[i]>ub,ub,df[i]))
```

```python
#pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE
c=SMOTE(random_state=1)
x,y=c.fit_resample(x,y)
```

```python
sns.countplot(x=y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)# train_size=0.7
```

```python
xtrain.shape,xtest.shape
```

```python
ytrain.shape,ytest.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
```

```python
xtest
```

```python
from sklearn.ensemble import RandomForestClassifier
r=RandomForestClassifier()

from sklearn.tree import DecisionTreeClassifier
d=DecisionTreeClassifier()

from sklearn.neighbors import KNeighborsClassifier
k=KNeighborsClassifier()

from sklearn.metrics import accuracy_score,classification_report,ConfusionMatrixDisplay
```

```python
a=[r,d,k]
for i in a:
    print (i)
    i.fit(xtrain,ytrain)
    ypred=i.predict(xtest)
    print(accuracy_score(ytest,ypred)*100)
    print(classification_report(ytest,ypred))
    print(ConfusionMatrixDisplay.from_predictions(ytest,ypred))
    plt.show()
```
